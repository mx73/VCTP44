```python
from qiskit import IBMQ
from qiskit import ClassicalRegister, QuantumRegister, QuantumCircuit
from qiskit import execute
from qiskit import Aer
from qiskit.tools.jupyter import *
#from qiskit.providers.ibmq import least_busy
# import basic plot tools
from qiskit.tools.visualization import plot_histogram, circuit_drawer
```

```python
provider0 = IBMQ.load_account()
provider1 = IBMQ.get_provider(hub='ibm-q', group='open', project='main')
backends = provider0.backends()
backend = provider0.get_backend('ibmqx4')
#backend = provider0.get_backend('ibmq_16_melbourne')

print("The backend is " + backend.name())
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
q = QuantumRegister(5)
c = ClassicalRegister(5)
qc = QuantumCircuit(q,c)

#state preparation
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.x(q[3])
qc.h(q[3])

#1st Grover iteration
qc.ccx(q[0],q[1],q[4])
qc.ccx(q[4],q[2],q[3])
qc.ccx(q[0],q[1],q[4])

qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[2])
qc.ccx(q[0],q[1],q[2])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])

#2nd Grover iteration
qc.ccx(q[0],q[1],q[4])
qc.ccx(q[4],q[2],q[3])
qc.ccx(q[0],q[1],q[4])

qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[2])
qc.ccx(q[0],q[1],q[2])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])

#3rd Grover iteration
qc.ccx(q[0],q[1],q[4])
qc.ccx(q[4],q[2],q[3])
qc.ccx(q[0],q[1],q[4])

qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[2])
qc.ccx(q[0],q[1],q[2])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])

#4th Grover iteration
qc.ccx(q[0],q[1],q[4])
qc.ccx(q[4],q[2],q[3])
qc.ccx(q[0],q[1],q[4])

qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[2])
qc.ccx(q[0],q[1],q[2])
qc.h(q[2])
qc.x(q[0])
qc.x(q[1])
qc.x(q[2])
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])

#Measure
qc.measure(q[0],c[0])
qc.measure(q[1],c[1])
qc.measure(q[2],c[2])
```
