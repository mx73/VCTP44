```python
import time
import numpy as np
from qiskit import IBMQ
from qiskit import ClassicalRegister, QuantumRegister, QuantumCircuit
from qiskit import execute, compile
from qiskit import Aer
from qiskit.tools.jupyter import *
from qiskit.providers.jobstatus import JobStatus, JOB_FINAL_STATES
# import basic plot tools
from qiskit.tools.visualization import plot_histogram, circuit_drawer
```

```python
IBMQ.load_accounts()
IBMQ.backends()
```

```python
backend = IBMQ.backends(name='ibmq_16_melbourne')[0]
print("The backend is " + backend.name())
simulator = Aer.get_backend('qasm_simulator')
```

# 6-8

```python
q1 = QuantumRegister(3)
c1 = ClassicalRegister(3)
qc1 = QuantumCircuit(q1, c1)
#    for k in range(n):
#        qc.h(q[k])
qc1.measure(q1, c1)
layout = [6, 7, 8]
print(layout)
```

```python
qc1.draw(output='mpl')
```

```python
start_time = time.time()
```

```python
job_1 = execute(qc1, backend=backend, initial_layout=layout, shots=8192)
counts_1 = job_1.result().get_counts(qc1)
```

```python
print("--- %s seconds ---" % round(time.time() - start_time), 2)
```

```python
print(counts_1)
```

```python
qb68 = {}
qb681 = sum((v for k, v in counts_1.items() if k.endswith('0')), 0)
qb682 = sum((v for k, v in counts_1.items() if k.endswith('1')), 0)
qb683 = sum((v for k, v in counts_1.items() if k.endswith(('00' , '01'))), 0)
qb684 = sum((v for k, v in counts_1.items() if k.endswith(('10' , '11'))), 0)
qb685 = sum((v for k, v in counts_1.items() if k.startswith('0')), 0)
qb686 = sum((v for k, v in counts_1.items() if k.startswith('1')), 0)

qb68.update([('qb60', qb681),
             ('qb61', qb682),
             ('qb70', qb683),
             ('qb71', qb684),
             ('qb80', qb685), 
             ('qb81', qb686)])
```

```python
plot_histogram(counts_1) #raw data
```

```python
plot_histogram(qb68)
```
